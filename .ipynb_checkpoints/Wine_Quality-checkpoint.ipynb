**Consideration of Statistical Values and Preliminary Reflections**

**pH Value** </br>
The pH value of a wine serves as an indicator of acidity and alkalinity in an aqueous solution, typically ranging from 0 (strongly acidic) to 14 (strongly alkaline). A lower pH value indicates a higher acidity level in the wine, making it more stable. Generally, wine pH ranges from 2.7 to 4.0, with the ideal range being between 2.9 and 3.9 (Wein-Plus GmbH n.d.a).

In this dataset, the maximum pH value of red wine is observed to be 4.0, slightly above the ideal range. This suggests that some wines in the dataset are more acidic. However, the average pH value falls within the ideal range at 3.31, indicating stable wine quality.

---

**Alcohol Content** </br>
The alcohol content in wine is measured in volume percent and indicates the amount of alcohol present relative to the total volume of the wine. This value depends on the ripeness of the grapes and typically ranges from 7 to 16 volume percent for red wine (K&U-Weinhalle n.d.).

In the present dataset, the average alcohol content of red wine is 10.42 volume percent, which falls below the specified range. 75% of the data exhibit an alcohol content of less than 11.10 volume percent, while the maximum value is 14.9 volume percent.

---

**Residual Sugar**

Residual sugar in wine is a byproduct of the natural fermentation process and primarily consists of fructose (fruit sugar), as glucose (grape sugar) ferments more quickly into alcohol and carbon dioxide. A small portion consists of non-fermentable sugars such as pentose. The residual sugar level is indicated on wine labels as dry, semi-dry, semi-sweet, sweet according to legal requirements (Wein-Plus GmbH n.d.b).

According to these guidelines, dry wines have residual sugar levels of up to 4 g/L, semi-dry wines up to 12 g/L, semi-sweet wines up to 45 g/L, and sweet wines over 45 g/L (Deutsches Weininstitut n.d.).

In the present dataset, the average residual sugar content is 2.54 g/dm³. 75% of the data have a residual sugar content of 2.60 g/dm³ or less, indicating a dry red wine. However, the maximum value of 15.5 g/dm³ suggests potential outliers in the dataset.

### Skewness

<div class="alert alert-block alert-info">
    Skewness measures the asymmetry of a probability distribution and indicates the direction in which the distribution deviates from horizontal symmetry.

- A distribution is positively skewed when the median is less than the mean.
- A distribution is symmetric when the median equals the mean.
- A distribution is negatively skewed when the median is greater than the mean. 

(Hemmerich n.d.)
</div> 

```python
df_skew = pd.DataFrame(index=df_red_wine.columns, columns=['mean', 'median', 'diff'])
df_skew['mean'] = df_red_wine.mean()
df_skew['median'] = df_red_wine.median()
df_skew['diff'] = df_red_wine.mean() - df_red_wine.median()
```

```python
df_skew.T
```

```python
def check_skewness_type(df):
    positive_skewness = {}
    negative_skewness = {}
    symmetric = []
    for label in df.columns:
        median = df[label].median()
        mean = df[label].mean()
        if median < mean:
            positive_skewness[label] = "Positively skewed"
        elif median == mean:
            symmetric.append(label)
        else:
            negative_skewness[label] = "Negatively skewed"
    
    return positive_skewness, negative_skewness


positive_skewness, negative_skewness = check_skewness_type(df_red_wine)

print("Positively skewed distributions:")
for label, skewness_type in positive_skewness.items():
    print(f"{label}")

print("\nNegatively skewed distributions:")
for label, skewness_type in negative_skewness.items():
    print(f"{label}")
```

<div class="alert alert-block alert-info">
    - If skewness is less than -1 or greater than 1, the distribution is highly skewed. 
    </br>
    - If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
    </br>
    - If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.
    </br>

(Klima 2021)
</div> 

```python
def check_skewness(df):
    print("Distribution is highly skewed for:")
    for label in df.columns:
        if (skew(df[label]) < -1) or (skew(df[label]) > 1):
            print(f"{label}: {skew(df[label]):.4f}")

    print("\nDistribution is moderately skewed for:")
    for label in df.columns:
        if (-1 <= skew(df[label]) <= -0.5) or (0.5 <= skew(df[label]) <= 1):
            print(f"{label}: {skew(df[label]):.4f}")

    print("\nDistribution is approximately symmetric for:")
    for label in df.columns:
        if -0.5 < skew(df[label]) < 0.5:
            print(f"{label}: {skew(df[label]):.4f}")

check_skewness(df_red_wine)
```

**Observation of the distribution** </br>
Based on the distributions of individual features, it is evident that a large portion is positive-skewed, particularly the features "residual sugar," "chlorides," "free sulfur," "total sulfur dioxide," and "sulphates." This suggests that low values occur frequently, as seen, for example, in the "residual sugar" feature. Additionally, this indicates the presence of some outliers, contributing to the significant distortion of the distributions. Conversely, the other features exhibit a nearly normal distribution of the data.

### A3: *"Which columns are correlated with the column 'quality'?"*

### Correlation

<div class="alert alert-block alert-info">
Correlation allows us to assess the linear relationship between two variables in terms of direction and strength.</br>

- <b>Positive correlation</b> occurs when the correlation coefficient ranges between 0 and 1, indicating that larger values of Variable A are associated with larger values of Variable B.
- <b>Negative correlation</b> occurs when the correlation coefficient ranges between -1 and 0, indicating that larger values of Variable A are associated with smaller values of B.

The <b>strength of correlation</b> can be classified as follows:
Magnitude of r | Strength of Relationship
-------------|--------------------------
0.0 < 0.1 | No correlation
0.1 < 0.3 | Weak correlation
0.3 < 0.5 | Moderate correlation
0.5 < 0.7 | Strong correlation 
0.7 < 1.0 | Very strong correlation

(DATAtab Team 2024)
</div>

```python
sns.set(rc={'figure.figsize': (14, 8)})
sns.heatmap(df_red_wine.corr(), cmap='rocket', annot=True, fmt='.2f', linewidths=2)
```

```python
df_red_wine.corr(numeric_only=True)['quality'].sort_values(ascending=False)
```

**Observations from the correlations in the dataset**

When examining the correlations between the feature quality and the other features, it is noticeable that the features alcohol, sulphates, citric acid, fixed acidity, and residual sugar exhibit a positive correlation. The strongest correlation exists between the features alcohol and quality, which is moderate at 0.4762. The remaining positively correlated features show a weaker correlation. 

In contrast, the features free sulfur dioxide, pH, chlorides, density, total sulfur dioxide, and volatile acidity show a negative correlation. The highest negative correlation is observed between the features volatile acidity and quality, which is also moderate with a magnitude of 0.3906.

From this, it can be inferred that there is a certain relationship between alcohol and quality, implying that alcohol positively influences quality. Conversely, volatile acidity appears to have a negative impact on the quality of the red wine.

### H1: *"Alcohol has a positive influence on the quality of the red wine."*

```python
# boxplot
fig, ax = plt.subplots(figsize=(20, 10))
ax.boxplot([df_red_wine[df_red_wine['quality'] == q]['alcohol'] for q in sorted(df_red_wine['quality'].unique())], showfliers=False)

ax.set_xlabel('Quality')
ax.set_ylabel('Alcohol')
ax.set_title('Boxplot of Alcohol Content by Quality')

ax.set_xticklabels(sorted(df_red_wine['quality'].unique()))

plt.show()
```

```python
# line plot
avg_alcohol_by_quality = df_red_wine.groupby('quality')['alcohol'].mean()

plt.figure(figsize=(10, 5))
plt.plot(avg_alcohol_by_quality.index, avg_alcohol_by_quality.values, marker='o', linestyle='-')

plt.xlabel('Quality')
plt.ylabel('Average Alcohol Content')
plt.title('Average Alcohol Content by Quality')

plt.show()
```

```python
sns.jointplot(x='quality',y='alcohol',data=df_red_wine, kind='reg')
```

**Observation from Boxplot and Line Plot** </br>
The visual representation of the relationship between alcohol content and quality was presented using both a box plot, jointplot and a line plot. It is evident from the graphics that the average alcohol content increases with the rising quality of the red wine.

### C: Influence of Alcohol and Volatile Acidity on Wine Quality

```python
quality_order = sorted(df_red_wine['quality'].unique())

plt.figure(figsize=(10, 6))

for quality in quality_order:
    data = df_red_wine[df_red_wine['quality'] == quality]
    plt.scatter(data['alcohol'], data['volatile acidity'], label=f'Quality {quality}', alpha=0.7)

plt.xlabel('Alcohol')
plt.ylabel('Volatile Acidity')
plt.title('Influence of Alcohol and Volatile Acidity on Wine Quality')

plt.legend()

plt.grid(True)
plt.show()
```

```python
g = sns.FacetGrid(df_red_wine, col="quality", col_wrap=3, height=4, aspect=1.5)
g.map(plt.scatter, "alcohol", "volatile acidity", alpha=0.5)

for ax in g.axes:
    quality = ax.get_title().split("=")[1].strip()
    mean_alcohol = df_red_wine[df_red_wine['quality'] == int(quality)]['alcohol'].mean()
    mean_volatile_acidity = df_red_wine[df_red_wine['quality'] == int(quality)]['volatile acidity'].mean()
    ax.plot(mean_alcohol, mean_volatile_acidity, marker='o', color='black')

g.set_axis_labels("Alcohol", "Volatile Acidity")
g.fig.suptitle('Relationship between Alcohol and Volatile Acidity by Quality', y=1.05)

plt.show()
```

**Summary of Exploratory Data Analysis**

- Prior to the exploratory data analysis, the assumption was made that pH level, alcohol content, and residual sugar might have an impact on the quality of red wine.
- During the exploratory data analysis, it was found that pH level and residual sugar do not show a significant correlation with wine quality.
- However, alcohol content does appear to have an influence on wine quality.
- It was also observed that the volatile acidity level may have a negative impact on wine quality.

Based on these insights from the dataset, initial hypotheses can be derived for further modeling work. For example, linear regression could be employed to predict wine quality using the available features.

Another consideration would be to apply a Random Forest model to predict, for instance, whether wine quality can be classified as 'good', 'medium', or 'poor'. This would enable the determination of a prediction class for each data point.
